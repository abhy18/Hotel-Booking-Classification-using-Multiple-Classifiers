```python
from sklearn.preprocessing import LabelEncoder
f=LabelEncoder()
columns=df.select_dtypes(include='object').columns
for i in columns:
  df[i]=f.fit_transform(df[i])


```

```python
df=df.drop(['company','agent','name','email','phone-number'],axis=1)
```

```python
df['country']=df['country'].fillna(df['country']).mean()
df['children']=df['children'].fillna(df['children']).mean()
```

```python
x=df.drop(['is_canceled'],axis=1)
y=df['is_canceled']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x,y)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred1=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred1,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf2=DecisionTreeClassifier(criterion='entropy')
clf2.fit(x,y)
```

```python
clf2.fit(x_train,y_train)
```

```python
y_pred2=clf2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred2,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf4=RandomForestClassifier(random_state=70,n_jobs=-1,max_depth=10,n_estimators=150,max_features=20,oob_score=True,criterion='entropy',max_leaf_nodes=40)
```

```python
clf4.fit(x_train,y_train)
```

```python
y_pred3=clf4.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred3,y_test)
```

```python
from sklearn.svm import SVC
clf3=SVC(kernel='rbf')
clf3.fit(x_train,y_train)
```

```python
y_pred4=clf3.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred4,y_test)
```
